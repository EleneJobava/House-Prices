```python
import nbformat

with open('model_inference.ipynb', 'r') as f:
    nb = nbformat.read(f, as_version=4)

for cell in nb.cells:
    if cell.cell_type == 'code':
        print('--- CODE ---')
        print(cell.source)
        print()
    elif cell.cell_type == 'markdown':
        print('--- MARKDOWN ---')
        print(cell.source)
        print()
```
